```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("instagram_data.csv")
data
```

```python
data.info
```

```python
print(data.isnull().sum())
```

```python
data.dropna(inplace=True)
```

```python
print(data.isnull().sum())
```

```python
data['comments'] = pd.to_numeric(data['comments'], errors='coerce')
```

```python
data.head()
```

```python
print(data.describe())
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['likes'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Likes')
plt.xlabel('Likes')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='followers', y='likes', data=data, color='orange')
plt.title('Likes vs Followers')
plt.xlabel('Followers')
plt.ylabel('Likes')
plt.show()
```

```python
data['likes'] = pd.to_numeric(data['likes'], errors='coerce')
```

```python
data['is_video'] = data['is_video'].astype('category')
```

```python
print(data.dtypes)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='is_video', y='likes', data=data, palette='Set2')
plt.title('Likes Distribution for Videos and Images')
plt.xlabel('Is Video')
plt.ylabel('Likes')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='owner_username', y='likes', data=data, estimator=sum, ci=None)
plt.title('Total Likes by Owner Username')
plt.xlabel('Owner Username')
plt.ylabel('Total Likes')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['comments'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Comments')
plt.xlabel('Number of Comments')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='followers', y='likes', data=data, color='orange')
plt.title('Likes vs Followers')
plt.xlabel('Followers')
plt.ylabel('Likes')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='is_video', data=data, palette='pastel')
plt.title('Count of Video vs. Image Posts')
plt.xlabel('Is Video')
plt.ylabel('Count')
plt.show()
```
